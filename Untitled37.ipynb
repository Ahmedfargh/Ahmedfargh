```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/stocks.csv')
df.head()
```

```python
df.Ticker.unique()
```

```python
nfx=df[df.Ticker=='NFLX']
apple=df[df.Ticker=='AAPL']
amazon=df[df.Ticker=='MSFT']
Goo=df[df.Ticker=='GOOG']
```

```python
def percents(open,close):
  return ((close-open)/open)*100
nfx=nfx.assign(percent_change=percents(nfx.Open,nfx.Close))
Goo=Goo.assign(percent_change=percents(Goo.Open,Goo.Close))
amazon=amazon.assign(percent_change=percents(amazon.Open,amazon.Close))
apple=apple.assign(percent_change=percents(apple.Open,apple.Close))
```

```python
len(nfx)
```

```python
percents=nfx.percent_change.tolist()
goog=Goo.percent_change.tolist()
amazon_perc=amazon.percent_change.tolist()
apple_perc=apple.percent_change.tolist()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(percents)
ax.plot(goog)
ax.plot(amazon_perc)
ax.plot(apple_perc)
ax.legend(['Netflix','Google','Amazon','Apple'])
ax.set_xlabel('Days')
ax.set_ylabel('Percent Change')
ax.set_title('NFLX Percent Change Over Time')
plt.show()
```

```python
def max_profit(percents):
  return max(percents)-min(percents)
goog_max=max_profit(goog)
amazon_max=max_profit(amazon_perc)
apple_max=max_profit(apple_perc)
nfx_max=max_profit(percents)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(['Google','Amazon','Apple','Netflix'],[goog_max,amazon_max,apple_max,nfx_max])
ax.set_title('Max Profit')
ax.legend(['Max Profit'])
ax.set_xlabel('Stocks')
ax.set_ylabel('Max Profit')
```

```python
nfx_avg=nfx.percent_change.mean()
Goo_avg=Goo.percent_change.mean()
amazon_avg=amazon.percent_change.mean()
apple_avg=apple.percent_change.mean()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(['Google','Amazon','Apple','Netflix'],[Goo_avg,amazon_avg,apple_avg,nfx_avg])
ax.set_title('Average Profit')
ax.legend(['Average Profit'])
ax.set_xlabel('Stocks')
ax.set_ylabel('Average Profit')
```

```python
goog=Goo.percent_change.std()
amazon_perc=amazon.percent_change.std()
apple_perc=apple.percent_change.std()
nfx_perc=nfx.percent_change.std()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(['Google','Amazon','Apple','Netflix'],[goog,amazon_perc,apple_perc,nfx_perc])
ax.set_title('Standard Deviation')
ax.legend(['Standard Deviation'])
ax.set
```

```python
def convidence(percents,z):
  std=np.std(percents)
  mean=np.mean(percents)
  return (std/np.sqrt(len(percents)))*z
```

```python
mean_nfx=convidence(nfx.percent_change,1.96)
mean_Goo=convidence(Goo.percent_change,1.96)
mean_amazon=convidence(amazon.percent_change,1.96)
mean_apple=convidence(apple.percent_change,1.96)
upper_nfx=np.mean(percents)+mean_nfx
lower_nfx=np.mean(percents)-mean_nfx
upper_Goo=np.mean(goog)+mean_Goo
lower_Goo=np.mean(goog)-mean_Goo
upper_amazon=np.mean(amazon_perc)+mean_amazon
lower_amazon=np.mean(amazon_perc)-mean_amazon
upper_apple=np.mean(apple_perc)+mean_apple
lower_apple=np.mean(apple_perc)-mean_apple
```

```python
# prompt: generate stacked bar

import pandas as pd
import matplotlib.pyplot as plt
# Creating the dataset
data = {'Upper': [upper_nfx, upper_Goo, upper_amazon, upper_apple],
        'Lower': [lower_nfx, lower_Goo, lower_amazon, lower_apple]}

# Creating the index
index = ['Netflix', 'Google', 'Amazon', 'Apple']

# Creating the dataframe
df = pd.DataFrame(data, index)

# Generating the stacked bar chart
ax = df.plot.bar(stacked=True, figsize=(10, 7))

# Set the title
ax.set_title('Stacked Bar Chart')

# Set the x and y labels
ax.set_xlabel('Stocks')
ax.set_ylabel('Confidence Interval')

# Display the chart
plt.show()
```

```python
def count_losses(data):
  count=0
  for i in data:
    if i<0:
      count+=1
  return count/len(data)
nfx_loss=count_losses(nfx.percent_change)
Goo_loss=count_losses(Goo.percent_change)
amazon_loss=count_losses(amazon.percent_change)
apple_loss=count_losses(apple.percent_change)
```

```python
fig,ax=plt.subplots(figsize=(12,8))
ax.bar(['Google','Amazon','Apple','Netflix'],[Goo_loss,amazon_loss,apple_loss,nfx_loss])
ax.set_title('Losses')
ax.legend(['Losses'])
ax.set_xlabel('Stocks')
ax.set_ylabel('Losses')
plt.show()
```
